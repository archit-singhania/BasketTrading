**Downloading the FINANCE DATA**

```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "META"]
```

```python
data = yf.download(
    tickers,
    start="2015-01-01",
    end="2025-01-01",
    auto_adjust=True
)
```

```python
prices = data["Close"]
```

```python
prices.to_csv("../data/basket_prices.csv")
```

**LOADING & CLEANING DATA**

```python
prices = pd.read_csv("../data/basket_prices.csv", index_col=0, parse_dates=True)
```

```python
print(prices.isnull().sum())
```

```python
prices = prices.ffill()
```

```python
prices.tail()
```

**VISUALIZING THE RAW PRICES**

```python
import matplotlib.pyplot as plt
```

```python
prices.plot(figsize=(12,6))
plt.title("Tech Stock Basket Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
